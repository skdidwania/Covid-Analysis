# <font color='green'>Resampling </font> data
> Here I have performed <b>weekly rolling</b> on the data.
 The  bins over which some aggregating functions are performed) are overlapping in this process.
 Resampling is a methodology of economically using a data sample to improve the accuracy and quantify the uncertainty of a   population parameter. Resampling methods, in fact, make use of a nested resampling method.
 The reason for using rolling mean:
 - Rolling means are quite handy for <b>assessing</b> the <b>trend</b> in our dataset. 
 <br>
 <font color='red'>Here we are computing  a 7-day rolling mean </font>

# <font color='green'>IRELAND</font>

```python
data_columns = ['retail_and_recreation_percent_change_from_baseline', 'grocery_and_pharmacy_percent_change_from_baseline', 'parks_percent_change_from_baseline', 'transit_stations_percent_change_from_baseline','workplaces_percent_change_from_baseline','residential_percent_change_from_baseline']
data_7d_rol_ireland = data_ireland[data_columns].rolling(window = 7, center = True).mean()
data_7d_rol_ireland
```
